```python
# DATA CLEANING
```

```python
#filling missing values
```

```python
data['Item_Weight']=data_train['Item_Weight'].fillna(data_train['Item_Weight'].mean())
data['Outlet_Size']=data_train['Outlet_Size'].fillna(data_train['Outlet_Size'].mode()[0])
```

```python
data['Item_Outlet_Sales'] = data['Item_Outlet_Sales'].replace(0, np.NaN)
data['Item_Outlet_Sales'].fillna(data['Item_Outlet_Sales'].mode()[0], inplace = True)
```

```python
data.isnull().sum()
```

```python
# FEATURE ENGINEERING
```

```python
#STAGE-1
data.groupby('Outlet_Type')['Item_Outlet_Sales'].mean()
```

```python
#STAGE-2
data['Item_Visibility']=data_train['Item_Visibility'].replace(0,data_train['Item_Visibility'].mean())
#data_test['Item_Visibility']=data_test['Item_Visibility'].replace(0,data_test['Item_Visibility'].mean())
```

```python
#STAGE-3
data.Item_Type_Combined = data.Item_Identifier.apply(lambda x: x[0:2])
data.Item_Type_Combined = data.Item_Type_Combined.map({'FD':'Food',
                                                             'NC':'Non-Consumable',
                                                             'DR':'Drinks'})

data['Item_Type_Combined']=data.Item_Type_Combined
data.Item_Type_Combined.value_counts()
```

```python
#STAGE-4
data.Item_Fat_Content=data.Item_Fat_Content.replace({'LF':'Low Fat','low fat':'Low Fat'})
data.Item_Fat_Content=data.Item_Fat_Content.replace({'reg':'Regular'})


data.Item_Fat_Content.value_counts()
```

```python
#droping unusual data columns
data.drop(['Item_Identifier','Outlet_Identifier','Outlet_Establishment_Year','Item_Type'],axis=1,inplace=True)
data
```

```python
#one hot encoding
```

```python
new_data=pd.get_dummies(data,columns=['Item_Fat_Content','Item_Type_Combined','Outlet_Size','Outlet_Location_Type','Outlet_Type'])
new_data
```
